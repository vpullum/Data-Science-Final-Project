## Load packages

```python
import pandas as pd
```

## Load dataset

```python
WFH_BurnOut = pd.read_csv("WFH_BurnOut.csv")
```

```python
WFH_BurnOut.head()
```

## Remove NULL values

```python
WFH_BurnOut1 = WFH_BurnOut.dropna()
```

```python
WFH_BurnOut1.head()
```

## Recoding string values to integer values

```python
def Gender (series):
    if series == "Female":
        return 0
    if series == "Male":
        return 1
WFH_BurnOut1["GenderR"] = WFH_BurnOut["Gender"].apply(Gender)
```

```python
def WFH_Setup_Available (series):
    if series == "No":
        return 0
    if series == "Yes":
        return 1
WFH_BurnOut1["WFH_Setup_AvailableR"] = WFH_BurnOut["WFH_Setup_Available"].apply(WFH_Setup_Available)
```

## Load secondary dataset-- La-Vie Borde

## Data Recoding

```python
WFH_Data.Devices.value_counts()
```

```python
def Devices_recode (series):
    if series == "Laptop;":
        return 0
    if series == "Laptop; Smartphone;":
        return 1
    if series == "Desktop;":
        return 2
    if series == "Desktop; Laptop;":
        return 3
    if series == "Desktop; Laptop; Tablet; Smartphone;":
        return 4
    if series == "Laptop; Tablet;":
        return 5
    if series == "Laptop; Tablet; Smartphone;":
        return 6
    if series == "Desktop; Laptop; Smartphone;":
        return 7
    if series == "Desktop; Laptop; Tablet; Smartphone; Other;":
        return 8
    if series == "Desktop; Smartphone;":
        return 9
    if series == "Smartphone;":
        return 10
    
WFH_Data["DevicesR"] = WFH_Data["Devices"].apply(Devices_recode)
```

```python
WFH_Data.head()
```

```python
WFH_Data.Obtain_access.value_counts()
```

```python
def Obtain_access_recode (series):
    if series == "Using VPN":
        return 0
    if series == "Using 3rd party cloud solutions (e.g. G Cloud, Microsoft 365)":
        return 1
    if series == "Direct access":
        return 2
    if series == "Other":
        return 3
    if series == "No Access":
        return 4
    if series == "No special provisions":
        return 5
    if series == "VPN access":
        return 6
    if series == "Remote connection to your workstation (e.g. TeamViewer)":
        return 7
    
WFH_Data["Obtain_accessR"] = WFH_Data["Obtain_access"].apply(Obtain_access_recode)
    
```

```python
WFH_Data.head()
```

```python
def New_apps_recode (series):
    if series == "No":
        return 0
    if series == "Yes":
        return 1
WFH_Data["New_appsR"]= WFH_Data["New_apps"].apply(New_apps_recode)
```

```python
def Productivity_recode (series):
    if series == "Agree":
        return 0
    if series == "Undecided":
        return 1
    if series == "Disagree":
        return 2
    if series == "Strongly agree":
        return 3
    if series == "Strongly disagree":
        return 4

WFH_Data["ProductivityR"] = WFH_Data["Productivity"].apply(Productivity_recode)
```
